```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

### Load the image using OpenCV

```python
im = cv2.imread('lenna.png')
```

```python
im.shape
```

### Convert the RGB image into a grayscale image

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

### Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

h_kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

### Do the filtering

```python
v_image = cv2.filter2D(im, 0, v_kernel)

h_image = cv2.filter2D(im, 0, h_kernel)

# Combine the vertical and horizontal edge images
image = cv2.addWeighted(v_image, 0.5, h_image, 0.5, 0)
```

### Display results

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(image, cmap='gray')
plt.title('Edge Image')

```
