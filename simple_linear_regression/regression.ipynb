# Simple Linear Regression

### Task: Predict student score given the number of hours spent by the student for studies
#### Input: Hours Studies
#### Output: Score

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("w1_student_scores.csv")

df.head(5)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(df['Hours'],df['Scores'], 'ro')
plt.title('Student Score Vs Hours Studied')
plt.xlabel('Hours')
plt.ylabel('Scores')

plt.show()
```

### Find the Coefficient

#### Calculate the mean of the input and the output

```python
xmean = np.mean(df['Hours'])
ymean = np.mean(df['Scores'])
```

#### Calculate the **Covariance** and the **Variance of the independent variable** of the each data point

```python
df['xycov'] = (df['Hours'] - xmean) * (df['Scores'] - ymean)
df['xvar'] = (df['Hours'] - xmean)**2
```

#### Calculate the **slope** and the **interception** to predict the best fited line for distribution

![alt text](image.png)

```python
slope = df['xycov'].sum()/df['xvar'].sum()

intercept = ymean - slope*xmean

print(slope,intercept)
```

### Make Prediction

```python
predict_scores = intercept + slope * df['Hours']
```

```python
plt.figure(figsize=(6,4))
plt.plot(df['Hours'], predict_scores)
plt.plot(df['Hours'], df['Scores'], 'ro')
plt.title('Actual VS Predicted')
plt.xlabel('Hours')
plt.ylabel('Scores')

plt.show()
```
